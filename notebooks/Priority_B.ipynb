```python
# defining the fields that we need to search for each ID
my_fields = ['name', 'permanently_closed']
```

# 2. Getting the formatted address for each store

```python
priority_b = pd.read_excel('Priority A_B_C_3.0_cleansed.xlsx', sheet_name='Priority_B_799 stores')
```

```python
priority_b['Country'] = 'Canada'
priority_b['Formatted_address'] = priority_b['Name'] + ',' + priority_b['Street_Address'] + ',' + priority_b['City__Town'] + ',' + \
priority_b['Postal code'] + ',' + priority_b['Country']
```

```python
priority_b.head()
```

# 3. Getting the unique ID for each of the stores using geocode api

```python
priority_b['id'] = 'na'
priority_b_loa = priority_b['Formatted_address']
for address in priority_b_loa:
    try:
        priority_b.loc[priority_b['Formatted_address'] == address, ['id']] = gmaps.geocode(address)[0]['place_id']
        # set a buffer of 0.3 sec when running on google cloud
        time.sleep(0.3)
    except:
        break
```

```python
priority_b.head()
```

# 4. Filtering the permanently_closed lists

```python
priority_b['Permanently_closed'] = False
```

```python
priority_b.head()
```

```python
priority_b_id_loa = priority_b['id']
```

```python
for id in priority_b_id_loa[:]:
    try:
        gmaps.place(place_id = id, fields = my_fields)['result']['permanently_closed']
        priority_b.loc[priority_b['id'] == id, ['Permanently_closed']] = True
    except:
        continue
```
